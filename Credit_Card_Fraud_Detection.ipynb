### Because we are having really few data points

#### 0 <--- NORMAL TRANSACTION

#### 1 <--- FRAUDULENT TRANSACTION 

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]

print(legit.shape)
print(fraud.shape)
```

### Under-Sampling

### Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

### Number of Fraudulent Transactions --> 492



```python
legit_sample = legit.sample(n=492)
```

```python
legit_sample.head()
```

```python
fraud.head()
```

### Concatenate legit_sample and fraud dataframes 

```python
new_dataset = pd.concat([legit_sample,fraud])
new_dataset.head()
```

```python
new_dataset.shape
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Class', data= new_dataset, palette="pastel")  
plt.title('Legit transactions and Fraudulent transactions')
plt.show()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Splitting the data into Features and Target

```python
X = new_dataset.drop ( columns = 'Class' , axis = 1) 
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

### SPLIT THE DATA INTO TRAIN AND TEST DATA 


```python
X_train , X_test , Y_train , Y_test = train_test_split( X , Y , test_size = 0.2 , stratify = Y , random_state = 2) 
```

```python
print (X_test)
```

```python
print( X.shape , X_train.shape , X_test.shape )
```

```python
print( Y.shape , Y_train.shape , Y_test.shape )
```

## Model training 

### Logistic Regression

```python
model = LogisticRegression() # create a varible as model 
```

### train the logistic regression model with training data

```python
model.fit ( X_train , Y_train )  
```

### To evaluate based on accuracy score ( Model Evaluation )

#### Accuracy on training data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score (X_train_prediction , Y_train ) 
```

```python
print ('Accuracy on training data : ' , training_data_accuracy ) 
```

#### Accuracy on test data

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score (X_test_prediction , Y_test) 
```

```python
print ('Accuracy on test data : ' , test_data_accuracy) 
```

### IF accuracy score on training data is extremely different from Accuracy on test data.

### IT can be more or less overfitting or underfitting 


```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(Y_test, X_test_prediction)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test, X_test_prediction))
```

## Confusion Matrix Explanation

Think of this as a **scoreboard** to see how many times your model got it right or wrong.

|               | **Predicted: 0** | **Predicted: 1** |
|---------------|------------------|------------------|
| **Actual: 0** | ✅ 97             | ❌ 2             |
| **Actual: 1** | ❌ 13             | ✅ 85            |

###  Breakdown:
- ✅ *97 times* — the model correctly said **"Not Fraud"**
- ❌ *2 times* — the model wrongly said **"Fraud"**, but it was actually **normal**
- ❌ *13 times* — the model wrongly said **"Not Fraud"**, but it was actually **fraud**
- ✅ *85 times* — the model correctly said **"Fraud"**

---

##  What Do These Terms Mean?

###  Precision
> *“Out of the times I said this is fraud, how often was I right?”*

- For **class 1 (fraud)**, it's **0.98** — *awesome!* ✅  
> _(When the model says it's fraud, it's almost always correct!)_

---

###  Recall
> *“Out of the actual frauds, how many did I catch?”*

- For **class 1**, it's **0.87** — *pretty good!*  
> _(Out of all frauds, the model caught 87% of them.)_

---

###  F1-Score
> *A combo of Precision & Recall — a final score to balance both.*

- Around **0.92** for both fraud and non-fraud — *great job!*

---

###  Support
> *Number of actual samples in each class.*

- **99** non-fraud, **98** fraud in your test data

---

##  Overall Scores:
- **Accuracy**: *92%* of the time, your model got it right
- **Macro Avg**: *Average for both classes* (good for imbalance)
- **Weighted Avg**: *Takes into account how many examples you had of each*
